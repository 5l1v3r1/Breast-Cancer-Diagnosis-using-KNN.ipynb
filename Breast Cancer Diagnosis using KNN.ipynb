```python
import numpy as np  
import pandas as pd 

# visualisation 
import matplotlib.pyplot as plt 

#reading the excel file
data = pd.read_csv(r"C:\Users\user\Desktop\data.csv")
data.head()
```

```python
#dropping columns which we dont need to classify
df = data.drop(['Unnamed: 32','id'],axis=1)
```

```python
#convert the diagnosis value of M and B to a numerical value
def diagnosis_value(diagnosis): 
    if diagnosis == 'M': 
        return 1
    else: 
        return 0
  
df['diagnosis'] = df['diagnosis'].apply(diagnosis_value) 
df
```

```python
#plotting the data
import seaborn as sns

sns.lmplot(x = 'radius_mean', y = 'texture_mean', hue = 'diagnosis', data = df) 
```

```python
#plotting the data
sns.lmplot(x ='smoothness_mean', y = 'compactness_mean',  
           data = df, hue = 'diagnosis') 
```

```python
#Splitting the data as train and test (test size = 33% of the whole data)
X = np.array(df.iloc[:, 1:-10]) 
y = np.array(df['diagnosis']) 

from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split( 
	X, y, test_size = 0.20, random_state = 42) 
```

```python
#appyling KNN algorithm
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 13) 
knn.fit(X_train, y_train) 
```

```python
#calculating KNN score
knn.score(X_test, y_test) 
```
